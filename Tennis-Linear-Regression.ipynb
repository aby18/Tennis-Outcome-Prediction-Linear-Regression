```python
import seaborn
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# loading and investigating the data
tennis = pd.read_csv('tennis_stats.csv')
print(tennis.head())
print(tennis.info())
```

```python
# Exploratory Data Analysis
# BreakPointsOpportunities vs Winnings
plt.scatter(tennis.Winnings, tennis.BreakPointsOpportunities, alpha = 0.4)
plt.title('BreakPointsOpportunities vs Winnings')
plt.xlabel('Winnings')
plt.ylabel('BreakPointsOpportunities')
plt.show()
plt.close()
```

```python
# ServiceGamesPlayed vs Wins
plt.scatter(tennis.Wins, tennis.ServiceGamesPlayed, alpha = 0.4)
plt.title('ServiceGamesPlayed vs Wins')
plt.xlabel('Wins')
plt.ylabel('ServiceGamesPlayed')
plt.show()
plt.close()
```

```python
# Aces vs Winnings
plt.scatter(tennis.Winnings, tennis.Aces, alpha = 0.4)
plt.title('Aces vs Winnings')
plt.xlabel('Winnings')
plt.ylabel('Aces')
plt.show()
plt.close()
```

```python
# BreakPointsOpportunities vs Wins
plt.scatter(tennis.Wins, tennis.BreakPointsOpportunities, alpha = 0.4)
plt.title('BreakPointsOpportunities vs Wins')
plt.xlabel('Wins')
plt.ylabel('BreakPointsOpportunities')
plt.show()
plt.close()
```

```python
# DoubleFaults vs Losses
plt.scatter(tennis.Losses, tennis.DoubleFaults, alpha = 0.4)
plt.title('DoubleFaults vs Losses')
plt.xlabel('Losses')
plt.ylabel('DoubleFaults')
plt.show()
plt.close()
```

```python
## Single Feature Linear Regression:

## BreakPointsOpportunities vs Wins
features = tennis[['BreakPointsOpportunities']]
outcome = tennis[['Wins']]

# Splitting data into training and test sets
features_train, features_test, outcome_train, outcome_test = train_test_split(features, outcome, train_size = 0.8, test_size = 0.2)

model = LinearRegression()
model.fit(features, outcome)
print(model.score(features_test, outcome_test))

prediction = model.predict(features_test)
plt.scatter(outcome_test, prediction, alpha = 0.4)
plt.title('Predicted vs Actual Values for BreakPointsOpportunities vs Wins')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.show()
plt.close()
```

```python
## ServiceGamesPlayed vs Wins
features = tennis[['ServiceGamesPlayed']]
outcome = tennis[['Wins']]

# Splitting data into training and test sets
features_train, features_test, outcome_train, outcome_test = train_test_split(features, outcome, train_size = 0.8, test_size = 0.2)

model = LinearRegression()
model.fit(features, outcome)
print(model.score(features_test, outcome_test))

prediction = model.predict(features_test)
plt.scatter(outcome_test, prediction, alpha = 0.4)
plt.title('Predicted vs Actual Values for ServiceGamesPlayed vs Wins')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.show()
plt.close()
```
